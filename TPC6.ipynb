Resumo: O código em python apresentado tem como objetivo a manipulação de polinómios. É pedido ao utilizador que introduza um polinómio e de seguida podem ser efetuadas uma série de operações como ver o grau do polinómio, calcular a derivada do polinómio, calcular o polinómio num ponto, tabelar e visualizar o polinómio num referencial.

```python
def criarTermo(coef,exp):
    return (coef,exp)


def criarPolinomio():
    return []


def inserirTermo(p,t):
    p.append(t)
    

def verTermo(t):
    c,e = t
    if e == 0:
        u=(str(c))
    elif c == 1:
        u=("x^" + str(e))
    elif e == 1:
        u=(str(c) + "x")
    else:
        u=(str(c) + "x^" + str(e))
    return u
        

def verPolinomio(p):
    politexto = " "
    if len(p) != 0:
        politexto = verTermo(p[0])
        for i in range(1,len(p)):
            if p[i][0] >= 0:
                politexto = str(politexto) + "+" + str(verTermo(p[i]))
            else:
                politexto = str(politexto) + str(verTermo(p[i]))
    return politexto
                

def grauPol(p):
    grau = p[0][1]
    for i in range(1,len(p)):
        if p[i][1] > grau:
            grau = p[i][1]
    return grau
            

def calcPol(p, x):
    valor = 0
    for (c,e) in p:
        valor = valor + (c*(x**e))
    return valor


def tabela(p,linhas):
    print("x" + "::" + verPolinomio(p))
    for l in range(0,linhas):
        print(str(l) + "::" + str(calcPol(p,l)))
        

def derivadaPol(p):
    derivada=[]
    for (c,e) in p:
        if e!= 0:
            derivada.append(((c*e), e-1))
    return derivada


def final(O):
    if O == "1":
        return appPol()
    else:
        print("Adeus!")

def critOrd(t):
    (c,e)=t
    return e

def ordenarPol(p):
    p.sort(key=critOrd , reverse = True)
    return p

def simplificarPol(p):
    paux = ordenarPol(p)
    simplificado = []
    while len(paux) > 1:
        res = paux[0]
        del paux[0]
        while res[1] == paux[0][1]:
            res = (res[0]+paux[0][0], res[1])
            del paux[0]
            simplificado.append(res)
    if paux[0][1] != simplificado[-1][1]:
        simplificado.append(paux[0])
    return simplificado

import matplotlib.pyplot as plt

def visualizarPol(p):
    x = [n for n in range(-100,100)]
    y = []
    for n in x:
        y.append(calcPol(p,n))
    plt.plot(x, y)
    eixoXx= [x[0]-1,x[-1]+1]
    eixoXy= [0,0]
    eixoYx=[0,0]
    eixoYy=[-200,200]
    plt.plot(eixoXx, eixoXy, color='red', linewidth=1)
    plt.plot(eixoYx, eixoYy, color='red', linewidth=1)
    plt.title('Polinómio p')
    plt.show()
 
        
        
print("Bem-vindo! Vamos trabalhar com polinómios!")
def appPol():
    n=int(input("Qual o número de termos do polinómio?: \n"))
    p=criarPolinomio()
    for i in range(n):
        coef=int(input("Introduza o coeficiente do termo:\n"))
        exp=int(input("Introduza o expoente do coeficiente do termo:\n"))
        inserirTermo(p, criarTermo(coef, exp))
    print(verPolinomio(p))
    a=int(input("""O que pretende fazer com o polinómio?
                1 - Ver o grau do polinómio
                2 - Calcular a derivada do polinómio
                3 - Calcular o polinómio num ponto
                4 - Simplificar o polinómio
                5 - Tabelar
                6 - Visualizar o gráfico do polinómio
                """))
    if a == 1:
        print("O grau do polinómio é", grauPol(p))
    if a == 2:
        print("A derivada do polinómio é", verPolinomio(derivadaPol(p)))
    if a == 3:
        x=int(input("Para que valor de x pretende calcular?:"))
        print(calcPol(p,x))
    if a == 4:
        print("O polinómio simplificado é", verPolinomio(simplificarPol(p)))
    if a == 5:
        linhas=int(input("Quantas linhas pretende que a tabela tenha?:\n"))
        tabela(p,linhas)
    if a == 6:
        visualizarPol(p)
    O = input("Pretende fazer uma nova operação com um novo polinómio (1) ou sair(2)?:\n")
    final(O)
        

appPol()
```
